##### Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

In pandas, the functions **loc** and **iloc** are used for indexing and selecting data from a DataFrame, but they have different ways of referencing data:

**1. loc:**

- The **loc** function is label-based indexing, which means you refer to the data based on the labels (row and column names).

- It accepts label-based indexing for both rows and columns.

- You provide the row label(s) and column label(s) as arguments.

- It can be used to select data using boolean arrays, slices, or even callable functions.

- Example: **df.loc[row_labels, column_labels]**

**2. iloc:**

- The **iloc** function is integer-based indexing, which means you refer to the data based on integer positions (row and column indices).

- It accepts integer-based indexing for both rows and columns.

- You provide the integer index(es) of the rows and columns as arguments.

- It can be used to select data using integer arrays, slices, or boolean arrays.

- Example: **df.iloc[row_indices, column_indices]**

##### Q6. What do you understand about the windows function in pandas and list the types of windows functions?

In pandas, a window function (also known as a rolling or moving function) is a mechanism to perform calculations on a specified window of data points that move along a series or DataFrame. This is particularly useful for tasks such as calculating moving averages, cumulative sums, or other aggregated statistics over a specific window of data.

The primary function used to apply window operations in pandas is the **.rolling()** function. It creates a "rolling view" over a specified window of data, allowing you to apply various aggregation functions.

**Here are some common types of window functions that can be applied using .rolling():**

**1. mean():** Calculate the rolling mean over the specified window.

**2. sum():** Calculate the rolling sum over the specified window.

**3. min():** Calculate the rolling minimum over the specified window.

**4. max():** Calculate the rolling maximum over the specified window.

**5. std():** Calculate the rolling standard deviation over the specified window.

**6. var():** Calculate the rolling variance over the specified window.

**7. apply():** Apply a custom function to each rolling window.

##### Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days, hours, and minutes using Pandas time delta. The program should prompt the user to enter the dates and display the result.

```python
import pandas as pd

print('Calculate the difference between two dates.\n')

date_1 = pd.to_datetime(input('Enter the date in YYYY-MM-DD format: '))
date_2 = pd.to_datetime(input('Enter the date in YYYY-MM-DD format: '))

time_difference = date_2-date_1

days = time_difference.days
hours = time_difference.total_seconds() / 3600
minutes =  time_difference.total_seconds() / 60

print(f'\nTotal days: {days} days, Total hours: {hours} hours, Total minutes: {minutes} minutes')
```

##### Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified column to a categorical data type. The program should prompt the user to enter the file path, column name, and category order, and then display the sorted data.

```python
import pandas as pd

file_path = input("Enter the file path of the CSV file: ")
column_name = input("Enter the column name to convert: ")
category_order = input("Enter the category order (comma-separated): ").split(',')

df = pd.read_csv(file_path)

df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

sorted_df = df.sort_values(by=column_name)

sorted_df
```

##### Q10. Write a Python program that reads a CSV file containing sales data for different products and visualizes the data using a stacked bar chart to show the sales of each product category over time. The program should prompt the user to enter the file path and display the chart.

```python
import pandas as pd

file_path = input("Enter the file path of the CSV file: ")

df = pd.read_csv(file_path)

df['Date'] = pd.to_datetime(df['Date'])

pivot_df = df.pivot(index='Date', columns='Category', values='Sales')

normalized_df = pivot_df.divide(pivot_df.sum(axis=1), axis=0)

normalized_df.plot(kind='bar', stacked=True, figsize=(10, 6), title='Normalized Sales by Product Category')
```

##### Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and displays the results in a table.

```python
import pandas as pd

file_path = input("Enter the file path of the CSV file: ")

df = pd.read_csv(file_path, index_col=None)

mean = df['Test Score'].mean()
median = df['Test Score'].median()
mode = df['Test Score'].mode()

summary = pd.DataFrame({'Statistic': ['Mean', 'Median','Mode'],
                        'Value': [mean, median, ', '.join(map(str, mode))]})

summary
```
